#üßπ **Pulizia e Preparazione dei Dati**

Le principali operazioni di pulizia e trasformazione includono:

**Caricamento e Rinomina Iniziale:**

Ogni file Excel √® stato caricato saltando le prime 4 righe (skiprows=4) per identificare correttamente le intestazioni.

Le prime due colonne di ogni dataset sono state rinominate in modo coerente: la prima in "Residenza" e la seconda con il nome specifico del contesto (es. "Destinazione", "Mezzo", "Alloggio").

**Gestione delle Celle Unite (ffill):**

√à stata applicata l'operazione .ffill() (fill forward) alla colonna "Residenza" di tutti i DataFrame. Questo passaggio √® stato cruciale per propagare i valori delle celle unite (es. "Nord", "Centro") alle righe successive che pandas avrebbe altrimenti letto come vuote (NaN).

**Trasformazione in Formato "Lungo" (melt_and_clean):**

√à stata definita una funzione generica melt_and_clean per standardizzare il formato dei dati. Questa funzione ha:

* Trasformato i DataFrame da un formato "wide" (dove gli anni erano colonne separate) a un formato "long" (con una colonna "Anno" e una colonna "Valore").

* Convertito la colonna "Valore" in tipo numerico, gestendo eventuali errori di conversione.

* Rimosso righe con valori mancanti (NaN) nella colonna "Valore".

* Pulito gli spazi bianchi dai nomi delle colonne e dai valori delle colonne chiave.

* Filtrato righe non pertinenti dalla colonna "Residenza" (es. quelle contenenti termini come "tipo", "territorio", "totale").

* Mantenuto solo i dati relativi all'anno "2023".

* Filtri Specifici per Dataset:

* Per tutti i dataset, sono state rimosse le righe dove la colonna "Residenza" conteneva il valore "Italia", in quanto considerato ridondante.

* Per Residenza e destinazione: Sono state rimosse le righe dove la colonna "Destinazione" conteneva i valori "Italia" o "Totale".

* Per Vacanze - residenza e mezzo di trasporto: Sono state rimosse le righe dove la colonna "Mezzo" conteneva il valore "Tutte le voci".

* Per Vacanze - residenza e tipo di alloggio: Sono state rimosse le righe dove la colonna "Alloggio" conteneva il valore "Totale".

**Salvataggio dei Dati Puliti:**

Infine, i tre DataFrame risultanti (df_dest_long, df_mezzi_long, df_alloggi_long), ora puliti e in formato "long", sono stati salvati come file CSV nella cartella Dataset_puliti su Google Drive, pronti per essere caricati e utilizzati in questo notebook per l'analisi e la visualizzazione.
